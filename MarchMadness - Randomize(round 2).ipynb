```python
team1score = NCAA2.team1_score
team2score = NCAA2.team2_score
```

```python
team1scorelist = []
for x in team1score:
    team1scorelist.append(x)
team2scorelist = []
for x in team2score:
    team2scorelist.append(x) 
```

```python
Team1_Output = []
for x in range(len(team1scorelist)):
    if team1scorelist[x]>team2scorelist[x]:
        Team1_Output.append(1)
    else:
        Team1_Output.append(0)
```

```python
NCAA2['team1_output']=Team1_Output
```
